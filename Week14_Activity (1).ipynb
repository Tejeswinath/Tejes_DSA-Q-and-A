```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X=data.drop('country',axis=1)
X.head()
```

```python
# Scalling using Standard Scaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(X)
```

```python
scaled_data
```

```python
scaled_data=pd.DataFrame(scaled_data,columns=X.columns)
```

```python
# Apply PCA for dimensionality reduction
```

```python
from sklearn.decomposition import PCA
pca=PCA(0.8)
x_pca=pca.fit_transform(scaled_data)
x_pca.shape
```

```python
x_pca
```

```python
pc=np.transpose(x_pca)
```

```python
pca_data=pd.DataFrame({'PCA_1':pc[0],'PCA_2':pc[1],'PCA_3':pc[2],'PCA_4':pc[3]})
```

```python
pca_data.head()
```

```python
pca_data.shape
```

```python
# Appling KMeans for Cluustering
df=pca_data
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    cluster=KMeans(n_clusters=i,init='k-means++',random_state=42)
    model=cluster.fit(df)
    wcss.append(model.inertia_)
    print(model.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel('No. of Clusters')
plt.ylabel('Inertia')
```

from the graph,the value of K corresponds to 4

```python
cluster=KMeans(n_clusters=4,init='k-means++',random_state=42)
model=cluster.fit(df)
```

```python
pd.Series(model.labels_).value_counts()
```

```python
new_details=pd.concat([df,pd.Series(model.labels_)],axis=1)
new_details.cloumns=['PCA_1','PCA_2','PCA_3','PCA_4','ClusterID']
new_details.head()
```

```python
new_details.columns
```

```python
new_details['Cluster_ID']=new_details[0]
```

```python
new_details.columns
```

```python
new_details.drop(0,axis=1,inplace=True)
new_details.head()
```

```python
new_details['Cluster_ID'].value_counts()
```

# Scatter plot b/w various PCA components

```python
# Scatter plot b/w PCA_1 and PCA_2
sns.scatterplot(x='PCA_1',y='PCA_2',hue='Cluster_ID',data=new_details,palette=['red','green','blue','black'])
```

```python
# Scatter plot b/w PCA_3 and PCA_4
sns.scatterplot(x='PCA_3',y='PCA_4',hue='Cluster_ID',data=new_details,palette=['red','green','blue','black'])
```

```python
details_merge=pd.merge(data,new_details,left_index=True,right_index=True)
details_merge.head()
```

```python
final_details=details_merge.drop(['PCA_1','PCA_2','PCA_3','PCA_4'],axis=1)
final_details.head()
```

```python
detail_analysis=final_details
child_mort_mean=pd.DataFrame(detail_analysis.groupby(['Cluster_ID']).child_mort.mean())
exports_mean=pd.DataFrame(detail_analysis.groupby(['Cluster_ID']).exports.mean())
health_mean=pd.DataFrame(detail_analysis.groupby(['Cluster_ID']).health.mean())
imports_mean=pd.DataFrame(detail_analysis.groupby(['Cluster_ID']).imports.mean())
income_mean=pd.DataFrame(detail_analysis.groupby(['Cluster_ID']).income.mean())
inflation_mean=pd.DataFrame(detail_analysis.groupby(['Cluster_ID']).inflation.mean())
life_expec_mean=pd.DataFrame(detail_analysis.groupby(['Cluster_ID']).life_expec.mean())
total_fer_mean=pd.DataFrame(detail_analysis.groupby(['Cluster_ID']).total_fer.mean())
gdpp_mean=pd.DataFrame(detail_analysis.groupby(['Cluster_ID']).gdpp.mean())
```

```python
final_analysis=pd.concat([pd.Series([0,1,2,3]),child_mort_mean,exports_mean,health_mean,imports_mean,income_mean,inflation_mean,life_expec_mean,total_fer_mean,gdpp_mean],axis=1)
```
